```python
w_df = pd.DataFrame.from_dict(nifty50_weights, orient='index', columns=['Weight'])
print(w_df)
```

```python
def force(comp1, comp2):
    weight1 = w_df.loc[comp1, 'Weight']
    weight2 = w_df.loc[comp2, 'Weight']
    correlation = df[comp1].corr(df[comp2])
    force_value = (weight1 * weight2)/((1-correlation)*2)
    return force_value
```

```python
force("RELIANCE.NS", "HDFCBANK.NS")
```

```python
def total_gravity(company, g):
    total_force = 0
    for comp in nifty_components:
        if comp != company:
            f = force(company, comp)
            total_force += f
    return total_force

    
```

```python
total_gravity("RELIANCE.NS", 9.26)
```

```python
G = 1
all_total_gravity = {}
for company in nifty_components:
    tg = total_gravity(company, G)
    all_total_gravity[company] = tg
    print(f"{company} Gravity Computed: {tg}")
    print("-----")
```

```python
all_total_gravity = dict(sorted(all_total_gravity.items(), key=lambda item: item[1], reverse=True))
print(all_total_gravity)
```
